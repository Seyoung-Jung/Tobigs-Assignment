② 이상치 확인

Z-score 값이 -3 미만 혹은 3 초과일 경우 이상치로 판명하고 평균값으로 대체

```python
# Z-score가 -3 미만 혹은 3 초과시 outlier로 판정
def outlier_zscore(data):
    zscore = (data-data.mean())/data.std()
    return np.where((zscore < -3) | (zscore > 3))[0]

# outlier를 None 값으로 바꾼 후 평균 대체
for i in [4,7,8,9,10,11,12,13]: ## 연속형 변수 칼럼 인덱스
    df.iloc[outlier_zscore(df.iloc[:,i]),i] = None
    df.iloc[:,i].fillna(df.iloc[:,i].mean(skipna=True), inplace=True)
    
df
```

# 2) feature engineering

```python
# infection_case에 'Church'가 포함되는지 묻는 변수 생성
df['church'] = df.infection_case.str.contains('Church')
df.church.sum()
```

wordcloud를 살펴보니 교회 감염이 많이 나오는 것을 파악하여 교회감염인지 아닌지 boolean 변수를 생성함. 117 observation 중 21개가 교회 감염인 것을 확인할 수 있음.

```python
# 확진자가 100명 이상인지 묻는 변수 생성
df['super_spread'] = df.confirmed > 100
df.super_spread.sum()
```

확진자가 100명 이상 나온 슈퍼 전파 케이스를 분류하기 위해 boolean 변수를 생성함. 117 observation 중 11개가 확진자 100명 이상의 슈퍼 전파 케이스인 것을 확인할 수 있음.

```python
df['university_exist'] = np.where(df.university_count==0,'exist','unexist')
```

해당 지역에 대학교가 존재하면 'exist', 존재하지 않으면 'unexist'인 str 변수를 생성함.

```python
df['children_than_elderly'] = (df.elementary_school_count + df.kindergarten_count)/df.elderly_population_ratio
```

노년층 대비 어린이가 많은 지역인지 파악하기 위해 초등학교 수와 유치원 수를 더한 후 노년층 인구비율로 나눠줌. 어린이층이 많을수록 값이 커지고 노년층이 많을수록 값이 작아짐.

```python
df['infection_category'] = df.infection_case.str.split().str[-1]
```
